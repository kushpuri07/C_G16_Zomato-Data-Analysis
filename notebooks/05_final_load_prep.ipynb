### Objective
This notebook covers:
1. Loading the cleaned dataset
2. KPI computation
3. Final data preparation for Tableau

Using cleaned dataset from 02_cleaning.ipynb

```python
import pandas as pd
import numpy as np
```

```python
# Load the cleaned dataset
df = pd.read_csv('/content/Zomato_Cleaned.csv')

print("Cleaned dataset loaded!")
print(f"Rows    : {df.shape[0]}")
print(f"Columns : {df.shape[1]}")
```

```python
# KPI 1 — Average Delivery Time
avg_delivery_time = df['Time_taken (min)'].mean().round(2)

print(f"Average Delivery Time : {avg_delivery_time} mins")
```

```python
# KPI 2 — Average Delivery Time by City
city_avg = df.groupby('City')['Time_taken (min)'].mean().round(2)

print("Average Delivery Time by City:")
print(city_avg)
```

```python
# KPI 3 — Average Delivery Time by Traffic Density
traffic_avg = df.groupby('Road_traffic_density')['Time_taken (min)'].mean().round(2)

print("Average Delivery Time by Traffic Density:")
print(traffic_avg)
```

```python
# KPI 4 — Average Delivery Time by Weather Conditions
weather_avg = df.groupby('Weather_conditions')['Time_taken (min)'].mean().round(2)

print("Average Delivery Time by Weather Conditions:")
print(weather_avg)
```

```python
# KPI 5 — Festival vs Non-Festival Average Delivery Time
festival_avg = df.groupby('Festival')['Time_taken (min)'].mean().round(2)

print("Average Delivery Time by Festival:")
print(festival_avg)
```

```python
# KPI 6 — Average Delivery Time by Vehicle Type
vehicle_avg = df.groupby('Type_of_vehicle')['Time_taken (min)'].mean().round(2)

print("Average Delivery Time by Vehicle Type:")
print(vehicle_avg)
```

```python
# KPI 7 — Top 10 Fastest Delivery Partners
top_partners = df.groupby('Delivery_person_ID')['Time_taken (min)'].mean().round(2).sort_values().head(10)

print("Top 10 Fastest Delivery Partners:")
print(top_partners)
```

```python
# KPI 8 — Order Volume by Hour
order_by_hour = df.groupby('Order_Hour')['ID'].count().reset_index()
order_by_hour.columns = ['Hour', 'Order_Count']

print("Order Volume by Hour:")
print(order_by_hour)
```

```python
# KPI 9 — Order Volume by Day of Week
order_by_day = df.groupby('Day_of_Week')['ID'].count().reset_index()
order_by_day.columns = ['Day', 'Order_Count']

print("Order Volume by Day of Week:")
print(order_by_day)
```

```python
# KPI 10 — Average Delivery Person Ratings by City
ratings_by_city = df.groupby('City')['Delivery_person_Ratings'].mean().round(2)

print("Average Delivery Person Ratings by City:")
print(ratings_by_city)
```

```python
# Save final dataset for Tableau(which is basically Zomato_Cleaned but we just renamed it as Zomato_Final)
df.to_csv('/content/Zomato_Final.csv', index=False)
print(f"Rows    : {df.shape[0]}")
print(f"Columns : {df.shape[1]}")
```
